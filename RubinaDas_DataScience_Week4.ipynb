**Rule Based Chatbot**



```python
intents = {
    "greeting": {
       "patterns": ["hi", "hello", "hey"],
       "responses": ["Hello!", "Hi there!", "Hey!"]
    },
    "price_query": {
        "patterns": ["price", "cost", "rate"],
        "responses": ["Please tell me the commodity name."]
    },
    "goodbye":{
        "patterns": ["bye", "exit"],
        "responses": ["Goodbye!", "See you later!"]
    }
}
```

```python
commodities = {"onion": 25, "potato": 20, "carrot": 50, "spinach": 15, "lemon": 5}
```

```python
import random

# intent detection function
def get_intent(user_input):
  user_input = user_input.lower()

  for intent, data in intents.items():
    for pattern in data["patterns"]:
      if pattern in user_input:
        return intent

  return "unknown"


# commodity detection function
def get_commodity(commodity):
  if commodity in commodities.keys():
    return True
  else:
    return False


# chatbot function
def chatbot():
  while True:
    user_input = input("User: ")
    intent = get_intent(user_input)

    if intent == "unknown":
      print("Bot: I didn't understand.")

    elif intent == "price_query":
      commodity = input(f"Bot: {''.join(intents[intent]["responses"])}").lower()

      if get_commodity(commodity):
        print(f"Bot: Price of {commodity} is Rs. {commodities[commodity]}.")
      else:
        print(f"Bot: I don't know the price of {commodity}.")

    elif intent == "goodbye":
      break

    else:
      response = random.choice(intents[intent]["responses"])
      print("Bot:", response)


chatbot()
```
