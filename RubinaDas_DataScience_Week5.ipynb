```python
def find_book(book_name):
    for category, books in book_database.items():
        for book_id, details in books.items():
            if details["Book Name"].lower() == book_name.lower():
                return category, book_id, details
    return None, None, None


def search_book():
    book_name = input("Enter book name: ")
    category, book_id, book = find_book(book_name)

    if book:
        if book["Availability"]:
            print(f"Book is available in {category}")
        else:
            print(f"Book is already issued to {book['Issued To']}")
    else:
        print("Book not found")
```

```python
def issue_book():
    try:
        student_id = input("Enter Student ID: ").upper()
        if student_id not in student_record.keys():
            print("Invalid Student ID")
            return

        book_name = input("Enter book name: ")
        category, book_id, book = find_book(book_name)

        if not book:
            print("Book not found")
            return

        if book["Availability"] == False:
            print("Book already issued or is unavailable at the moment")
            return
        else:
            # Issue book
            book["Availability"] = False
            book["Issued To"] = student_id

            student_record[student_id]["Issued Books"].append(book_id)
            student_record[student_id]["Due Date"][book_id] = book["Due Days"]

            print("Book issued successfully")

    except Exception as e:
        print("Error occurred:", e)
```

```python
def return_book():
    try:
        student_id = input("Enter Student ID: ").upper()
        if student_id not in student_record.keys():
            print("Invalid Student ID")
            return

        book_id = input("Enter Book ID: ").upper()

        if book_id not in student_record[student_id]["Issued Books"]:
            print("This book was not issued to this student")
            return

        # Find book
        for category, books in book_database.items():
            if book_id in books:
                book = books[book_id]
                break

        days_late = int(input("Enter number of late days: "))

        fine = days_late * book["Fine Per Day"]

        # Update records
        book["Availability"] = True
        book["Issued To"] = None

        student_record[student_id]["Issued Books"].remove(book_id)
        del student_record[student_id]["Due Date"][book_id]
        student_record[student_id]["Fine"] += fine

        print(f"Book returned successfully. Fine: ₹{fine}")

    except ValueError:
        print("Please enter valid number for days")
    except Exception as e:
        print("Error occurred:", e)
```

```python
def view_student():
    student_id = input("Enter Student ID: ").upper()

    if student_id in student_record:
        student = student_record[student_id]
        print("\nStudent Details")
        print("Name:", student["Name"])
        print("Issued Books:", student["Issued Books"])
        print("Due Dates:", student["Due Date"])
        print("Fine:", student["Fine"])
    else:
        print("Student not found")
```

```python
# ---------------- CHATBOT LOOP ---------------- #

def chatbot():
    print("Welcome to Library Assistant Chatbot")

    while True:
        print("\nChoose an option:")
        print("1. Search Book")
        print("2. Issue Book")
        print("3. Return Book")
        print("4. View Student Details")
        print("5. Exit")

        choice = input("Enter choice: ")

        if choice == "1":
            search_book()

        elif choice == "2":
            issue_book()

        elif choice == "3":
            return_book()

        elif choice == "4":
            view_student()

        elif choice == "5":
            print("Exiting chatbot")
            break

        else:
            print("nvalid choice. Try again.")
```

```python
# ---------------- RUN CHATBOT ---------------- #
chatbot()
```
